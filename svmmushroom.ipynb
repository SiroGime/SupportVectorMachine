### Se cambiará el nombre de las columnas.

```python
data = data.rename(columns={
    "p":"target",
    "x":"cap-shape",
    "s":"cap-surface",
    "n":"cap-color",
    "t":"bruises?",
    "p.1":"odor",
    "f":"gill-attachment",
    "c":"gill-spacing",
    "n.1":"gill-size",
    "k":"gill-color",
    "e":"stalk-shape",
    "e.1":"stalk-root",
    "s.1":"stalk-surface-above-ring",
    "s.2":"stalk-surface-below-ring",
    "w":"stalk-color-above-ring",
    "w.1":"stalk-color-below-ring",
    "p.2":"veil-type",
    "w.2":"veil-color",
    "o":"ring-number",
    "p.3":"ring-type",
    "k.1":"spore-print-color",
    "s.3":"population",
    "u":"habitat"
})
```

### Se cambiará el nombre de los valores de la variable objetivo/independiente ("target").

```python
data["target"] = data["target"].replace({"e":"edible",
                                         "p":"poisonous"})
```

```python
data.head()
```

### Se convertirá las variables categóricas en variables numéricas

```python
from sklearn.preprocessing import LabelEncoder   
encoder = LabelEncoder()

for var in data.columns.drop("target"):
    data[var] = encoder.fit_transform(data[var])
```

```python
data.head()
```

```python
data.isna().sum()
```

### Se dividirán los datos en datos de entrenamiento y datos de pruebas y se normalizarán las variables numéricas.

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(["target"], axis=1)
Y = data["target"]
```

```python
X.astype("float")
```

```python
X = (X-X.min())/(X.max()-X.min())
```

```python
X.head()
```

```python
X["veil-type"] = 0.0
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,
                                            test_size=0.3,
                                            random_state=0)
```

```python
len(x_train), len(x_test), len(y_train), len(y_test)
```

## Construcción del modelo.
* #### Se eligirán los mejores parametros para el modelado del support vector machine realizando una validación cruzada

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
parameters = [
    {
        "kernel": ["rbf"],
        "gamma": [1e-4,1e-3,1e-2,0.1,0.2,0.5],
        "C": [1,10,100,1000]
    },
    {
        "kernel":["linear"],
        "C": [1,10,100,1000]
    },
    {
        "kernel":["poly"],
        "C": [1,10,100,1000]
    }
]
```

```python
clf = GridSearchCV(SVC(decision_function_shape="ovr"),
                  param_grid=parameters, cv=5)
clf.fit(x_train, y_train)
```

```python
clf.best_params_
```

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

* #### Los mejores parametros para support vector machine son kernel radial basis function, C=1 y gamma=0.5 .

## Evaluación del modelo

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
```

```python
mat = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(mat.T, square=True, annot=True, fmt="d",
           cbar=True,xticklabels=y_test.unique(),
           yticklabels=y_test.unique())
```
